accept marks and display result

```python
nm=input('Enter student name ')
sub=input('Enter subject ')
mks=int(input('Enter marks '))

if mks>=35:
  print(f'Congratulations {nm}')
  print(f'you are pass in {sub}')
else:
  print(f'Sorry {nm}')
  print(f'you are fail in {sub}')
```

```python
amount=float(input('Enter purchase amount '))

if amount>20000:
  print('13% discount applied')
  discount=amount*13/100
else:
  print('9% discount applied')
  discount=amount*9/100

print(f'discount on amount {amount} is {discount}')
```

```python
n=int(input('Enter a number '))
if n>0:
  print('number is positive')
elif n<0:
  print('number is negative')
else:
  print('number is zero')
```
