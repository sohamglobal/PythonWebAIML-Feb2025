**Arithmetic Operators**

```python
a=9
b=13
print(a+b)
print(a-b)
print(a*b)
print('%.2f' %(a/b))
print(b%a)
print(a**2)
print(b//a)
```

**Relational operators**

```python
a=9
b=13
print(a==b)
print(a>b)
print(a>=b)
print(a<b)
print(a<=b)
print(a!=b)
```

**Logical operators**

```python
print((13>9) and (19<25))
print((13<9) or (19<25))
print(not 45>100)
```

**Assignment operators**

```python
x=5
x+=3
x-=4
x*=2
x/=3
x%=2
print(x)
```

**Bitwise operators**

0000,0001,0010,0011,0100,0101

```python
#bitwise and operator
a=5
b=3
result=a & b
print(result)
```

```python
#bitwise or operator
a=5
b=3
result=a | b
print(result)
```

```python
#bitwise xor operator
a=5
b=3
result=a ^ b
print(result)
```

**Membership operator**

```python
nm='microsoft'
print('p' in nm)
print('p' not in nm)
```

**Indentity operators**

```python
x=y=9
print(x is y)
print(x is not y)
```

```python
print(isinstance(13,int))
print(isinstance('praffull',float))
```
