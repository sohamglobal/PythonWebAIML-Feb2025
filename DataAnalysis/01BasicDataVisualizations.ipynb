```python
import pandas
import matplotlib.pyplot as plt
```

```python
data={
    'year':[2019,2020,2021,2022,2023,2024],
    'sales':[345,296,189,380,335,396]
}
```

```python
df=pandas.DataFrame(data)
print(df)
```

**Line Chart**

```python
plt.plot(df['year'],df['sales'])
plt.title('laptop yearly sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

**Bar chart**

```python
plt.bar(df['year'],df['sales'],color='skyblue')
plt.title('yearly mobile sales')
plt.xlabel('Year')
plt.ylabel('no of mobiles sold')
plt.show()
```

**Horizontal bar chart**

```python
plt.barh(df['year'],df['sales'],color='purple')
plt.title('yearly mobile sales')
plt.xlabel('no of mobiles sold')
plt.ylabel('Year')
plt.show()
```

**Scatter chart**

```python
plt.scatter(df['year'],df['sales'],color='green')
plt.plot(df['year'],df['sales'],color='orange')
plt.title('soham laptop shoppe')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

**Pie chart**

```python
sizes=df['sales']
labels=df['year']

plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title('sales distribution')
plt.show()
```

```python
dic={
    'status':['single','married','divorced'],
    'count':[5300,1700,3000]
}
df1=pandas.DataFrame(dic)
print(df1)
```

```python
sizes=df1['count']
lbl=df1['status']

plt.pie(sizes,labels=lbl,autopct='%1.1f%%',colors=['green','blue','red'])
plt.show()
```

**Box plot**

```python
import numpy as np
data=np.random.randn(100)

plt.boxplot(data)
plt.title('Box plot chart')
plt.ylabel('random values')
plt.show()
```

**Histogram**

```python
data=np.random.randn(1000)
plt.hist(data,bins=30,color='blue',edgecolor='black')
plt.show()
```
