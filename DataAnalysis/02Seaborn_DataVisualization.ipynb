```python
import pandas
import seaborn
from matplotlib import pyplot
```

```python
data={
    'player':['Salah','Trent','Darwin','Gakpo','Jota','Luis'],
    'goals':[29,3,5,10,6,13]
}
```

```python
df=pandas.DataFrame(data)
print(df)
```

```python
seaborn.lineplot(x='player',y='goals',data=df)
pyplot.title('Liverpools scorers')
pyplot.show()
```

```python
seaborn.barplot(x='player',y='goals',data=df)
pyplot.title('Liverpools scorers')
pyplot.show()
```

```python
seaborn.barplot(x='goals',y='player',data=df)
pyplot.title('Liverpools scorers')
pyplot.show()
```

```python
seaborn.scatterplot(x='player',y='goals',data=df)
pyplot.title('Liverpools scorers')
pyplot.show()
```

```python
seaborn.histplot(df['goals'])
pyplot.title('Liverpools scorers')
pyplot.show()
```
