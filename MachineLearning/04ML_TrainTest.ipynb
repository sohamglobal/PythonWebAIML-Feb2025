```python
import pandas
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/moviechoice.csv"
df=pandas.read_csv(url)
print(df)
```

```python
features=df[['gender','age','locality']]
labels=df['genre']
```

**take 80% data for training and 20% for testing**

```python
features_train,features_test,labels_train,labels_test=train_test_split(features,labels,test_size=0.2)
```

```python
model=DecisionTreeClassifier()
model.fit(features_train,labels_train)
```

```python
result=model.predict(features_test)
print('predicted/recommended by the model')
print(result)
print('actual results')
print(labels_test)
```
