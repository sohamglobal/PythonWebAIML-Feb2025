# Unsupervised learning using K-Means algorithm

```python
import pandas
from sklearn.cluster import KMeans
from matplotlib import pyplot
```

```python
data={
    'Player':['Aarya','Abhinav','Ajinkya','Akanksha','Anam','Apurva'],
    'Runs':[450,40,300,20,390,60],
    'Wickets':[2,13,5,15,3,10]
}
```

```python
df=pandas.DataFrame(data)
print(df)
```

```python
X=df[['Runs','Wickets']]
```

```python
km=KMeans(n_clusters=3,random_state=42)
df['Cluster']=km.fit_predict(X)
```

```python
print('Clustered Data : ')
print(df)
```

```python
pyplot.scatter(df['Runs'],df['Wickets'],c=df['Cluster'],cmap='viridis',s=100)
pyplot.xlabel('Runs Scored')
pyplot.ylabel('Wickets Taken')
pyplot.title('Cricket Players Clustering')
pyplot.colorbar(label='Cluster')
pyplot.grid(True)

for i,player in enumerate(df['Player']):
  pyplot.annotate(player,(df['Runs'][i]+5, df['Wickets'][i]+0.5))

pyplot.show()
```
